## Consultas em SQL

### Número de livros lançados após 1 de janeiro de 2000

```python
#Conjunto de dados
books
```

```python
#Realizando a consulta
query = '''
SELECT
    SUM(z.livros) AS "qtde_livros"
FROM    (SELECT
        COUNT(t.title) AS "livros"
        FROM  (SELECT
                books.title
                FROM books
                WHERE publication_date > '2000-01-01') AS t
                GROUP BY
                    t.title) AS z
'''
books_public = pd.io.sql.read_sql(query, con = engine)
books_public
```

Acima nota-se a tabela gerada com as informações dos livros publicados a partir de 01-01-2000, onde a data de publicação vai de 05-01-2000 até 31-03-2020.

```python
#Realizando a consulta
query='''
SELECT
    books.book_id,
    books.title,
    (SELECT 
        COUNT(reviews.book_id)
    FROM reviews
    WHERE reviews.book_id = books.book_id) AS qtd_avaliações,
    (SELECT 
        AVG(ratings.rating)
    FROM ratings
    WHERE ratings.book_id = books.book_id) AS class_media
FROM books
ORDER BY
    qtd_avaliações DESC;
'''

books_avg_rating = pd.io.sql.read_sql(query,con= engine)
books_avg_rating
```

### Editora que lançou o maior número de livros com mais de 50 páginas.

```python
#Conjunto de dados
books
```

```python
#Conjunto de dados
publishers
```

```python
#Realizando a consulta
query ='''
SELECT
    publisher,
    COUNT(*) AS "count_publishers"
FROM (
SELECT
    p.publisher
FROM
    books b
JOIN
    publishers p USING (publisher_id)
WHERE
    b.num_pages > 50) AS PU
GROUP BY
    publisher
ORDER BY
    count_publishers DESC
LIMIT 1;
'''

books_publ = pd.io.sql.read_sql(query,con= engine)
books_publ.head(10)
```

```python
books_publ['count_publishers'].sum()
```

Pelas tabelas geradas acima, verifica-se a distribuição dos livros por editora, onde é possivel notar que as TOP10 editoras ( das 340 editoras listadas) respondem por cerca de 22% de toda a publicação do período.

### Autor com a média mais alta de classificação de livros: livros com pelo menos 50 classificações.

```python
#Conjunto de dados
books.head()
```

```python
#Conjunto de dados
authors.head()
```

```python
#Conjunto de dados
ratings.head()
```

```python
#Realizando a consulta
query ='''
SELECT
    b.author_id,
    AVG(t.avg_rating) AS avg_rating
FROM
    books b
INNER JOIN(
SELECT
    r.book_id,
    COUNT(r.book_id) AS "qtde_rating",
    AVG(r.rating) AS "avg_rating"
FROM
    ratings r
GROUP BY
    book_id
HAVING
    COUNT(r.book_id) > 50) t ON t.book_id = b.book_id
GROUP BY
    author_id
ORDER BY
    AVG(t.avg_rating) DESC
'''

books_z = pd.io.sql.read_sql(query,con= engine)
books_z
```

```python
#Realizando a consulta
query ='''
SELECT
    a.author_id,
    a.author,
    z.avg_rating
FROM
    authors a
INNER JOIN(
SELECT
    b.author_id,
    AVG(t.avg_rating) AS avg_rating
FROM
    books b
INNER JOIN(
SELECT
    r.book_id,
    COUNT(r.book_id) AS "qtde_rating",
    AVG(r.rating) AS "avg_rating"
FROM
    ratings r
GROUP BY
    book_id
HAVING
    COUNT(r.book_id) > 50) t ON t.book_id = b.book_id
GROUP BY
    author_id
ORDER BY
    AVG(t.avg_rating) DESC) z ON z.author_id = a.author_id
ORDER BY
    z.avg_rating DESC
LIMIT 1;
'''

books_v = pd.io.sql.read_sql(query,con= engine)
books_v
```

Conform visto acima, após aplicar o filtro de livros com mais de 50 classificações, e a partir disso calcular a média, nota-se que os 3 primeiros da listas possuem média de classificação com valores bem próximos.

No enunciado é solicitado o cálculo do número médio de avaliações (review) dos usuários que realizaram mais de 50 classificações (rating). Porém há dois problemas: no conjunto de dados reviews não há o campo rating (notas) dos usuários; esse campo de encontra no conjunto de dados ratings. Outro ponto é que, agrupando o cojunto de dados reviews por username e contando os valores, nota-se que o número máximo de avaliações é 29 (número bem menor do que o pedido no enunciado, que é acima de 50). 

```python
#Realizando a consulta
query='''
SELECT
    ra.username,
    COUNT(ra.username) AS "qtde_username",
    AVG(ra.rating) AS "avg_rating"
FROM
    ratings ra
GROUP BY
    ra.username
HAVING
    COUNT(ra.username) > 50
ORDER BY
    COUNT(ra.username) DESC
'''

book_user = pd.io.sql.read_sql(query, con=engine)
book_user
```

Conforme visto acima e explicado anteriormente, tem-se acima os usuários que fizeram mais de 50 avaliações e suas respectivas médias. Nota-se que há empate entre primeiro e segundo lugares e terceiro e quarto lugares em quantidade de livros classificados. Percebe-se também que há usuários com média de avaliações maiores do que o segundo lugar, porém com quantidades menores de livros avaliados.
